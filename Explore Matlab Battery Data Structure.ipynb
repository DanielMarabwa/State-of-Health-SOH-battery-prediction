In this code, we go through battery number "B0005" to explore it's data structure. This will help us build a correct data cleaning process for the remaining 28 batteries.

```python
import scipy.io
import numpy as np
import pandas as pd
Battery_name = "B0005"
mat = scipy.io.loadmat(Battery_name)
battery = mat[Battery_name]
```

Date and time in matlab date vectir format. Convert this into Date.

```python
battery["cycle"][0,0][0,0]["time"][0].tolist()
```

```python
from datetime import datetime
time = battery["cycle"][0,0][0,0]["time"][0].tolist()
Battery_cycle_time = datetime(year = int(time[0]), #Year
        month = int(time[1]), #Month
        day = int(time[2]), #Day
        hour = int(time[3]), #Hour
        minute = int(time[4]), #Min
        second = int(time[5])) #second
Battery_cycle_time
```

```python
dict_time_of_measurement = {}
for i in range (battery["cycle"][0,0].size):
    type_of_measurement = str(battery["cycle"][0,0][0,i]["type"][0])
    time = battery["cycle"][0,0][0,i]["time"][0].tolist()
    Battery_cycle_time = datetime(year = int(time[0]),
        month = int(time[1]),
        day = int(time[2]),
        hour = int(time[3]),
        minute = int(time[4]),
        second = int(time[5]))
    #year = int(time[0])
    #month = int(time[1])
    #day = int(time[2])
    #hour = int(time[3])
    #minute = int(time[4])
    #second = (time[5])
    dict_time_of_measurement.update({i:[type_of_measurement,Battery_cycle_time]})
    #print(Battery_cycle_time)
df_Battery_measurements = pd.DataFrame.from_dict(dict_time_of_measurement, orient='index')
df_Battery_measurements.columns = ['Type of measurement','Battery cycle time']
df_Battery_measurements
```

We want to confirm that each iteration is one after the other and we use the sort_values to confirm that the index position aligns with the battery test timestamp.

```python
df_Battery_measurements.sort_values(by='Battery cycle time', ascending=True)
```
